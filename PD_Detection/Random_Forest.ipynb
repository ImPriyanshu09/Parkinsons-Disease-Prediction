```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('parkinsons.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.drop(['status','name'],axis = 1)
y = df['status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_x_train = scaler.fit_transform(X_train)
scaled_x_test = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
```

```python
x_train = pca.fit_transform(scaled_x_train)
x_test = pca.transform(scaled_x_test)
variance = pca.explained_variance_ratio_
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
n_estimators = [64,100,128,200]
max_features = [2,3,4]
bootstrap = [True,False]
oob_score = [True,False]
param_grid = {'n_estimators':n_estimators,'max_features':max_features,'bootstrap':bootstrap,'oob_score':oob_score}
```

```python
rfc = RandomForestClassifier()
grid = GridSearchCV(rfc,param_grid)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid_preds = grid.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,grid_preds))
```

```python
accuracy_score(y_test,grid_preds)
```

```python
rfc_model = RandomForestClassifier(max_features=2,n_estimators = 100,oob_score = True)
```

```python
rfc_model.fit(X_train,y_train)
```

```python
predicts = rfc_model.predict(X_test)
```

```python
print(classification_report(y_test,predicts))
```

```python
accuracy_score(y_test,predicts)
```
